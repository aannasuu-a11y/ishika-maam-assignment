```python
%%writefile app.py
import streamlit as st

# Rule-based logic dictionary
rules = {
    "hi": "Hello! I am HealthyBot. How can I help you stay healthy today?",
    "water": "You should aim for about 2-3 liters of water a day!",
    "sleep": "Getting 7-9 hours of sleep is crucial for recovery and focus.",
    "exercise": "Try to get at least 30 minutes of moderate activity daily.",
    "bye": "Stay healthy! Goodbye!"
}

def get_response(user_input):
    user_input = user_input.lower()
    for key in rules:
        if key in user_input:
            return rules[key]
    return "I'm not sure about 그. Try asking about water, sleep, or exercise!"

# Streamlit UI
st.title("🍎 HealthyBot: Your Rule-Based Assistant")
st.markdown("Type a keyword like **'water'**, **'sleep'**, or **'hi'** to start.")

if "messages" not in st.session_state:
    st.session_state.messages = []

# Display chat history
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

# User input
if prompt := st.chat_input("Ask me something..."):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    # Generate response
    response = get_response(prompt)

    with st.chat_message("assistant"):
        st.markdown(response)
    st.session_state.messages.append({"role": "assistant", "content": response})
```
